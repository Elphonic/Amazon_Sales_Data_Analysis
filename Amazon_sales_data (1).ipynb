## DATA ANALYSIS (Amazon Sales Data)
### Description:
This dataset contains information on 1K+ Amazon products, including their ratings, reviews, and other details.

### Features:
- **product_id**: Unique identifier for each product  
- **product_name**: Name of the product  
- **category**: Category of the product  
- **discounted_price**: Discounted price of the product  
- **actual_price**: Actual price of the product  
- **discount_percentage**: Percentage of discount for the product  
- **rating**: Rating of the product (1-5)  
- **rating_count**: Number of people who voted for the Amazon rating  
- **about_product**: Description about the product  
- **user_id**: ID of the user who wrote the review  
- **user_name**: Name of the user who wrote the review  
- **review_id**: ID of the user review  
- **review_title**: Short review  
- **review_content**: Long review  
- **img_link**: Image link of the product  
- **product_link**: Official website link of the product  

This dataset is useful for analyzing product trends, customer reviews, and pricing strategies on Amazon.

```python
import pandas as pd

# 'latin1' encoding
df_ams = pd.read_csv("amazon.csv", encoding="latin1")
df_ams.head()
```

```python
# List of numerical columns
num_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

# Cleaning and converting
for col in num_cols:
    df_ams[col] = df_ams[col].astype(str).str.replace(',', '')  # Remove commas
    df_ams[col] = df_ams[col].str.replace('%', '')  # Remove percentage signs
    df_ams[col] = pd.to_numeric(df_ams[col], errors='coerce')  # Convert to float

# Verify changes
print(df_ams.dtypes)
```

Q3  What is the distribution of discounted prices vs. actual prices?

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Use 'fill=True' instead of 'shade=True'
sns.kdeplot(df_ams['actual_price'], label="Actual Price", fill=True, color='r')
sns.kdeplot(df_ams['discounted_price'], label="Discounted Price", fill=True, color='b')

plt.xlabel("Price")
plt.ylabel("Density")
plt.title("Distribution of Discounted vs Actual Prices")
plt.legend()
plt.savefig("Amazon_Sales_Visualization/Distribution of Discounted vs Actual Prices.png", dpi=100, bbox_inches='tight')
plt.show()
```

# 📊 Insights from the Distribution of Discounted vs. Actual Prices

The density plot (Kernel Density Estimate - KDE) visualizes how **actual prices (red)** and **discounted prices (blue)** are distributed.

## 1️⃣ Major Observations

### **Both distributions are highly skewed to the right (positively skewed)**
- Most products have low prices, and only a few have extremely high prices.
- There are long tails extending toward higher values, indicating the presence of expensive items.

### **Peak of the Discounted Prices (Blue) is Higher than Actual Prices (Red)**
- The highest density (most frequent price range) is lower for discounted prices compared to actual prices.
- This suggests that many products get significant discounts, shifting their prices downward.

### **Discounted Prices are More Concentrated at Lower Values**
- The blue curve is **more concentrated on the left**, meaning many products are being sold at much lower prices than their actual value.
- This could indicate **deep discounts** on expensive items or more budget-friendly items making up the majority of sales.

### **Actual Prices Have a More Spread-Out Distribution**
- The red curve extends further, showing that actual prices vary greatly.
- Some high-value products remain expensive even after discounts.

Q6 What are the most popular product keywords

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re

# Extract product names
product_names = df_ams['product_name'].dropna().astype(str)

# Tokenize words: Remove special characters, split by space
words = []
for name in product_names:
    words.extend(re.findall(r'\b\w+\b', name.lower()))  # Convert to lowercase & extract words

# Remove common stopwords (expand this list if needed)
stopwords = {'for', 'with', 'and', 'the', 'of', 'to', 'in', 'on', 'by', 'a', 'an'}
filtered_words = [word for word in words if word not in stopwords]

# Count word frequencies
word_counts = Counter(filtered_words)
top_words = word_counts.most_common(10)  # Top 10 keywords

# Convert to DataFrame for visualization
df_keywords = pd.DataFrame(top_words, columns=['Keyword', 'Count'])

# Plot the most popular keywords
plt.figure(figsize=(10, 5))
sns.barplot(x='Count', y='Keyword', data=df_keywords, hue='Keyword', palette='viridis', legend=False)

plt.xlabel("Count")
plt.ylabel("Keyword")
plt.title("Most Popular Product Keywords")
plt.savefig("Amazon_Sales_Visualization/Most Popular Product Keywords.png", dpi=100, bbox_inches='tight')
plt.show()
```

Q8. What is the correlation between discounted_price and rating?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation
correlation = df_ams[['discounted_price', 'rating']].corr()

print("Correlation between Discounted Price and Rating:")
print(correlation)
```

# Correlation between Discounted Price and Rating

## 🔢 1️⃣ Understanding the Correlation Value
Correlation ranges from **-1 to 1**:

- **+1.0** → Perfect **positive** correlation (as one increases, the other also increases).
- **-1.0** → Perfect **negative** correlation (as one increases, the other decreases).
- **0.0** → **No correlation** (no relationship between the variables).

The correlation value is **0.126**, which indicates a **weak positive correlation**.

---

## 📊 2️⃣ Interpretation of This Result
- Since **0.126** is **close to 0**, it means there is a **very weak relationship** between the **discounted price** and **rating**.
- This suggests that **discounted price has little influence on product ratings**.
- In simple terms, **increasing or decreasing the discount price does not strongly impact how customers rate the product**.

---

## 💡 3️⃣ Business Insight
Since the correlation is weak, customers may rate products based on factors **other than price**, such as:

- ✅ **Quality of the product**  
- 🌟 **Brand reputation**  
- ⚙️ **Features and performance**  
- 📦 **Customer service and delivery experience**

```python
import os
from nbformat import v4

# Set the correct folder path for Amazon Sales visualizations
folder_path = "Amazon_Sales_Visualization"

# Ensure the folder exists
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Create a new Jupyter notebook
nb = v4.new_notebook()

# Load all saved images and add markdown + image cells
cells = []

# Check if the folder contains any images
image_files = sorted([f for f in os.listdir(folder_path) if f.endswith(".png")])

if not image_files:
    print(f"❌ No images found in the '{folder_path}' folder. Make sure to save plots first.")
else:
    for filename in image_files:
        image_path = f"{folder_path}/{filename}"
        
        # ✅ Formatting the title for each visualization
        markdown_text = f"### Visualization: {filename.replace('.png', '').replace('_', ' ').title()}"
        img_markdown = f"![{filename}]({image_path})"
        
        # Add markdown cell with title
        cells.append(v4.new_markdown_cell(markdown_text))
        # Add markdown cell with image
        cells.append(v4.new_markdown_cell(img_markdown))

    # Add cells to the notebook
    nb.cells.extend(cells)

    # Save notebook
    output_notebook = "Amazon_Sales_Visualizations_Summary.ipynb"
    with open(output_notebook, "w", encoding="utf-8") as f:
        f.write(v4.writes(nb))

    print(f"✅ All visualizations saved in '{output_notebook}' successfully!")
```
